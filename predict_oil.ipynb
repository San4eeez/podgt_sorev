```python
import pandas as pd

from statsmodels.tsa.arima.model  import ARIMA

df = pd.read_csv('test.csv',index_col=0)

df = df.dropna()

df['name_int'] = df['cty_name'].factorize()[0]


datka = df.select_dtypes(include=['int64','float64'])
# # 
# # 
# # 
# # datka.corr()
# 
# 
to_corr = datka[['gas_price_nom','year','gas_price_2000','oil_price_2000','sovereign']]
# 
# 

to_corr.set_index(df['year'])

model = ARIMA(to_corr['gas_price_nom'], order=(1,2,0))
model_fit = model.fit()

predictions = model_fit.forecast(steps=5)
print(predictions)




# 
# 
# model = ARIMA(to_corr['gas_price_nom'], order=(1,2,0))
# model_fit = model.fit()
# 
# predictions = model_fit.forecast(steps=5)
# 
# print(predictions)
```
